# Simple Data Analysis
This notebook loads a CSV and plots a line chart. Rebuilt in 2025 from earlier exercises.


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('weather.csv')
df
```

```python
plt.figure()
plt.plot(df['day'], df['temp_c'])
plt.title('Temperature by Day')
plt.xlabel('Day')
plt.ylabel('Temp C')
plt.show()
```
